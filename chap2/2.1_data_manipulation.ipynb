```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([
    [2.0, 1, 4, 3], 
    [1, 2, 3, 4], 
    [4, 3, 2, 1]])
torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
```

```python
X == Y
```

```python
X.sum()
```

```python
a = torch.arange(3).reshape(3, 1)
b = torch.arange(2).reshape(1, 2)
a, b
```

```python
a + b
```

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print(f'id(Z) = {id(Z)}')
Z[:] = X + Y
print(f'id(Z) = {id(Z)}')
```

```python
before = id(X)
X += Y
id(X) == before
```

```python
import numpy as np
A = np.array([1, 2, 3])
B = torch.tensor(A)
print((A, B))
B[1] = 4
print((A, B))
B.numpy()
```

```python
A = np.array([1, 2, 3])
B = torch.from_numpy(A)
print((A, B))
B[1] = 4
print((A, B))
B.numpy()
```
